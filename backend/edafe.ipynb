### Basic Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Loading the dataset

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
```

### Data cleaning

```python
dataset[dataset.isnull().any(axis = 1)]
```

```python
dataset.shape
```

### The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region
1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Region Dataset" 

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
```

```python
dataset[['Region']] = dataset[['Region']].astype(int)
```

```python
dataset.head(5)
```

```python
## Removing the null values
dataset = dataset.dropna().reset_index(drop = True)
```

```python
dataset.head(5)
```

```python
# Remove the 122nd row
dataset = dataset.drop(122).reset_index(drop = True)
```

```python
dataset.loc[122]
```

```python
# Fix spcaes in column names
dataset.columns = dataset.columns.str.strip()
```

```python
dataset.columns
```

```python
dataset.head(5)
```

```python
# Change the required column as integer datatype
dataset[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = dataset[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
dataset.head(5)
```

```python
# Changing the columns to float datatype
objects = [features for features in dataset.columns if dataset[features].dtypes == 'O']

for i in objects:
    if i!= 'Classes':
        dataset[i]=dataset[i].astype(float)
```

```python
dataset.info()
```

```python
dataset.to_csv("Algerian_forest_fires_dataset_cleaned.csv", index = False)
```

### EDA

```python
#Drop day month and year
df_copy = dataset.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy
```

```python
df_copy['Classes'].value_counts()
```

```python
plt.style.use('seaborn-v0_8')
```

```python
df_copy.hist (bins=50, figsize=(20,15))
plt.show()
```

```python
## Percentage of piechart

perc = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Plotting pie chart

classLabels = ["Fire", "Not Fire"]
plt.figure(figsize=(12, 7))
plt.pie(perc, labels=classLabels, autopct="%1.1f%%")
plt.title("Pie Chart")
```

```python
## Correlaion

df_corr = df_copy.corr()
df_corr
```

```python
sns.heatmap(df_corr)
```

```python
## Box plots
sns.boxplot(df_copy['FWI'], color = "g")
```
